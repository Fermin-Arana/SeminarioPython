**EJERCICIO 1.** Escribir una función que retorne los usuarios que accedieron al menos una vez  al material de clases  los días lunes, y las fechas en las que lo hicieron. Esta función recibe un parámetro opcional que indica el orden en que se incluyen las fechas: "**A**", si se ordenan en forma ascendente, "**D**" si es descendente o **None** si no se aplica ningún orden en particular. Por defecto la función no ordena.

**NOTA:** la información sobre el material de clases se obtiene verificando el texto "Página: Material de las clases" en la columna "Contexto del evento" del dataset.

**EJERCICIO 2.** Utilizando la función anterior, escribir un programa que muestre dos listados correspondientes a la actividad  de **los 5 usuarios** que más veces accedieron al material de clases los días lunes: uno en el cual  se muestren las fechas en orden descendente  y otro tal cual como aparecen en el dataset

### Solución a analizar

**EJERCICIO 1**

```python
import csv
from pathlib import Path
from datetime import datetime
def Accesos(csv_reader ,orden = None):
    accesos_lunes =[]
    for columna in csv_reader:
        hora=columna[0]
        usuario=columna[2]
        fecha = datetime.strptime(hora,"%d/%m/%Y")
        if fecha.weekday() == 0:
            accesos_lunes.append(hora,usuario,fecha)
    if orden == "A" :
        accesos_lunes.sort(l =lambda x:x[0])
    elif orden == "D":
        accesos_lunes.sort(l =lambda x:x[0], reverse=True)
for hora , usuario , fecha in accesos_lunes:
    print(f"El usuario {usuario} accedio el lunes {hora}") 
```

```python
file_route= Path('Archivos')/ "logs_catedra.csv"
file = open(file_route, "r")
csv_reader=csv.reader(file,delimiter=",")
orden=input('Ingresa el Orden a imprimir Los usuarios que accedieron el dia lunes: "A" para ascendente "D" para descendente o "Enter" solamente para sin orden')
Accesos(csv_reader,orden)
file.close
    
```

2.1. **Mostrá una solución dada por una IA a este ejercicio y respondé:**

```python
import csv
from pathlib import Path
from datetime import datetime
from collections import Counter

def obtener_accesos_lunes(csv_reader, orden=None):
    """
    Ejercicio 1: Filtra accesos de lunes al material de clase.
    Retorna un diccionario: { usuario: [lista_de_fechas] }
    """
    accesos = {}
    
    # Saltamos el encabezado si existe
    next(csv_reader, None)
    
    for fila in csv_reader:
        # Estructura supuesta: columna 0=Fecha, 1=Contexto, 2=Nombre usuario
        # Según enunciado: Filtrar "Página: Material de las clases" en Contexto
        try:
            fecha_str = fila[0]
            contexto = fila[1]
            usuario = fila[2]
            
            if "Página: Material de las clases" in contexto:
                fecha_dt = datetime.strptime(fecha_str.split(',')[0], "%d/%m/%Y")
                
                if fecha_dt.weekday() == 0:  # 0 es Lunes
                    if usuario not in accesos:
                        accesos[usuario] = []
                    accesos[usuario].append(fecha_str)
        except (ValueError, IndexError):
            continue

    # Aplicar orden a las listas de fechas de cada usuario
    if orden in ["A", "D"]:
        reversa = (orden == "D")
        for usuario in accesos:
            accesos[usuario].sort(key=lambda x: datetime.strptime(x.split(',')[0], "%d/%m/%Y"), reverse=reversa)
            
    return accesos

def ejercicio_2():
    path_archivo = Path('archivos') / "logs_catedra.csv"
    
    if not path_archivo.exists():
        print("Error: El archivo no existe.")
        return

    # Procesar para los dos listados requeridos
    with open(path_archivo, "r", encoding="utf-8") as file:
        # Lista para el dataset original (sin orden)
        reader_raw = csv.reader(file)
        datos_original = obtener_accesos_lunes(reader_raw, orden=None)
        
        # Resetear puntero para la segunda lectura o copiar el diccionario
        file.seek(0)
        reader_desc = csv.reader(file)
        datos_desc = obtener_accesos_lunes(reader_desc, orden="D")

    # Obtener los 5 usuarios con más accesos
    top_5 = sorted(datos_original.items(), key=lambda x: len(x[1]), reverse=True)[:5]
    usuarios_top = [u[0] for u in top_5]

    print("--- TOP 5 USUARIOS: FECHAS EN ORDEN DESCENDENTE ---")
    for usuario in usuarios_top:
        print(f"Usuario: {usuario} | Fechas: {datos_desc[usuario]}")

    print("\n--- TOP 5 USUARIOS: FECHAS TAL CUAL EL DATASET ---")
    for usuario in usuarios_top:
        print(f"Usuario: {usuario} | Fechas: {datos_original[usuario]}")

# Llamada al programa
# ejercicio_2()
```
